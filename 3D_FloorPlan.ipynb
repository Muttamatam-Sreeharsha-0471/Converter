#**Provided Input image path and Output image path**

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2

input_image_path = 'Input-Image.png'
output_image_path = 'Output_image.png'

input_image = Image.open(input_image_path)

rotated_image = input_image.rotate(-90, expand=True)

rotated_image_path = 'Rotated_Image.png'
rotated_image.save(rotated_image_path)

rotated_image.show()
```

#**Detected the Edges of Input image and Roatated that img top view**

```python

rotated_image_cv = cv2.cvtColor(np.array(rotated_image), cv2.COLOR_RGB2BGR)
gray_image = cv2.cvtColor(rotated_image_cv, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray_image, threshold1=50, threshold2=150)

edges_image_path = 'Edges_Image.png'
cv2.imwrite(edges_image_path, edges)

plt.imshow(edges, cmap='gray')
plt.title('Edge Detected Image')
plt.show()
```

#**Converted that image into Binary Img**

```python
inverted_edges = cv2.bitwise_not(edges)

_, binary_image = cv2.threshold(inverted_edges, 128, 255, cv2.THRESH_BINARY)

binary_image_path = 'Output_image.png'
cv2.imwrite(binary_image_path, binary_image)
```
